```python
import matplotlib.pyplot as plt

# Data
models = ['CLRerNet-DLA34', 'CondLSTR', 'CLRNet', 'FENet v2', 'FENet v1', 'GANet']
original_scores = [81.12, 80.77, 80.47, 80.19, 80.15, 79.63]
improved_scores = [87.46, 85.17, 84.87, 83.99, 83.85, 82.83]

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(models, original_scores, marker='o', linestyle='-', color='b', label='Original F1 Score')
plt.plot(models, improved_scores, marker='s', linestyle='--', color='r', label='Improved F1 Score')

# Annotate numerical values and differences
for i, model in enumerate(models):
    # Numerical values
    plt.text(i, original_scores[i] - 0.5, f'{original_scores[i]:.2f}', color='b', ha='center', fontsize=9)
    plt.text(i, improved_scores[i] + 0.5, f'{improved_scores[i]:.2f}', color='r', ha='center', fontsize=9)
    # Difference
    difference = improved_scores[i] - original_scores[i]
    plt.text(i, (original_scores[i] + improved_scores[i]) / 2, f'+{difference:.2f}', 
             color='green', ha='center', fontsize=9, fontweight='bold')

# Labels and formatting
plt.xlabel('Models')
plt.ylabel('F1 Score')
plt.title('Original vs. Improved F1 Score on CuLane Dataset')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
